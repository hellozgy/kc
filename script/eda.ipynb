```python
import pandas as pd
import numpy as np
```

```python
data_path = '../input/'
```

```python
train_df = pd.read_csv(data_path + 'train.csv')
test_df = pd.read_csv(data_path + 'test.csv')
```

```python
train_df.head()
```

```python
def judge_pure_english(keyword):  
    count = sum(ord(c) < 128 for c in keyword)
    rate = count / len(keyword)
    return rate < 0.4
    
```

```python
train_df[train_df['comment_text'].apply(judge_pure_english)]
```
